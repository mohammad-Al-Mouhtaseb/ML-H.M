```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.impute import SimpleImputer
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

import warnings
warnings.filterwarnings('ignore')
```

# Read CSV

```python
train=pd.read_csv('train.csv')
test=pd.read_csv('test.csv')
```

# Preprocess

```python
train.drop(['Name',"Ticket"],axis=1,inplace=True)

le1 = LabelEncoder()
train['Sex'] = le1.fit_transform(train['Sex'])

train.dropna(subset=["Embarked"], inplace=True)
train.reset_index(drop=True)

age_avg=train['Age'].value_counts().mean
train['Age'].replace("",age_avg,inplace=True)
train['Age']=train['Age'].astype('float64')

le2 = LabelEncoder()
train['Cabin'] = le2.fit_transform(train['Cabin'])
train = train.fillna(method='ffill') # for Cabin

le3 = LabelEncoder()
train['Embarked'] = le3.fit_transform(train['Embarked'])

train.info()
train.head()
```
